```python
perc = 80.0
min_count =  int(((100-perc)/100)*crashes.shape[0] + 1)
mod_df = crashes.dropna( axis=1, 
                thresh=min_count)
```

```python
mod_df = mod_df.drop(['RD_NO', 'REPORT_TYPE', 'DATE_POLICE_NOTIFIED', 'STREET_NO', 'STREET_DIRECTION', 'STREET_NAME'], axis = 1)
```

```python
percent_missing = mod_df.isnull().sum() * 100 / len(mod_df)
missing_value_df = pd.DataFrame({'column_name': mod_df.columns,
                                 'percent_missing': percent_missing})

pd.set_option('display.max_rows', None)
missing_value_df
```
